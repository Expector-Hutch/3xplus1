# 3x+1问题
这个问题有很多别名，都是以~~被它难倒~~的数学家命名的，由于别名太多了，所以大家干脆叫它3x+1问题了。   
该问题的内容为：有一个整数，如果它是整数，就把它除以2；如果它是一个奇数，就把它乘三再加一。  
无论如何，这个程序总会变成4,2,1,4,2,1...的循环。  
好消息是，3x+1问题已经在2022年1月被解决啦！

"$3x+1$"的数学表述如下：


设 $n \in \mathbb{Z}$, 且
$$
f(n) = \left\{ \begin{array}{ll} \frac{n}{2}, & 2 \mid n \\ 3n+1, & 2 \nmid n \end{array} \right.
$$
令
$$
f_1(n)=f(n),f_2(n)=f(f_1(n)),\cdots,f_k(n)=f(f_{k-1}(n))
$$
则存在有限正整数 $m \in \mathbb{N}$, 使得 $f_m(n)=4, f_{m+1}=2, f_{m+2}=1$.

用一个函数去验证3x+1问题，并时刻把运算结果输出。下面举个栗子~

```python
def threexp1(x: int):
    while x != 1:
        x = x // 2 if x % 2 == 0 else 3 * x + 1
        print(x)
    print('OK')
```
